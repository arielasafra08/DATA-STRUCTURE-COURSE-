##Tugas 6 - List dan Perulangan

```python
nilai = []
total = 0

for i in range(5):
    n = int(input("Masukkan nilai: "))
    nilai.append(n)
    total += n

rata = total / len(nilai)

print("Total =", total)
print("Rata-rata =", rata)
```

> Analisis

Program menggunakan list untuk menyimpan lima nilai mahasiswa dan memanfaatkan perulangan untuk memasukkan data secara berulang. Setelah seluruh nilai tersimpan, program menghitung total nilai dan rata-rata nilai. Hasil yang diperoleh menunjukkan bahwa list dan perulangan dapat digunakan bersama untuk memproses sejumlah data dengan lebih efisien.

> Kesimpulan

Penggunaan list dan perulangan mempermudah proses pengumpulan serta pengolahan data dalam jumlah banyak tanpa harus menuliskan kode yang berulang.

##Tugas 7 - Fungsi

```python
def luas_persegi_panjang(panjang, lebar):
    return panjang * lebar

p = float(input("Masukkan panjang: "))
l = float(input("Masukkan lebar: "))

print("Luas =", luas_persegi_panjang(p, l))
```

Hasil percobaan menunjukkan bahwa fungsi dapat digunakan untuk memisahkan logika perhitungan dari program utama.

> Kesimpulan

##Tugas 11 - Gabungan Semua Konsep

```python
def hitung_rata(data):
    return sum(data) / len(data)

nilai = []

for i in range(5):
    n = int(input("Masukkan nilai: "))
    nilai.append(n)

rata = hitung_rata(nilai)

if rata >= 60:
    status = "Lulus"
else:
    status = "Tidak Lulus"

print("Rata-rata =", rata)
print("Status =", status)
```

> Analisis

Program mengintegrasikan variabel, list, perulangan, fungsi, dan percabangan dalam satu aplikasi sederhana. Nilai mahasiswa disimpan ke dalam list, kemudian dihitung rata-ratanya menggunakan fungsi. Setelah itu dilakukan evaluasi menggunakan percabangan untuk menentukan status kelulusan. Hasil percobaan menunjukkan bahwa beberapa konsep dasar pemrograman dapat bekerja secara terpadu untuk menyelesaikan suatu permasalahan.

> Kesimpulan

Penguasaan konsep dasar pemrograman sangat penting karena setiap konsep saling berkaitan dan dapat digabungkan untuk membangun program yang lebih kompleks.

##Tugas 12 - Simulasi Kasir

```python
barang = []

for i in range(3):
    harga = int(input("Masukkan harga barang: "))
    barang.append(harga)

total = sum(barang)

if total >= 100000:
    diskon = total * 0.10
else:
    diskon = 0

bayar = total - diskon

print("Total Belanja =", total)
print("Diskon =", diskon)
print("Total Bayar =", bayar)
```

> Analisis

Program menerima input harga beberapa barang kemudian menghitung total belanja. Jika total belanja mencapai syarat tertentu, program memberikan diskon sebesar 10%. Setelah diskon dihitung, program menampilkan jumlah pembayaran akhir. Hasil percobaan menunjukkan bahwa percabangan dapat digunakan untuk menerapkan aturan transaksi secara otomatis.

> Kesimpulan

Konsep perhitungan dan percabangan dapat diterapkan dalam sistem kasir sederhana untuk membantu proses transaksi menjadi lebih cepat dan akurat.

##Tugas 14 - Faktorial dengan Fungsi

```python
def faktorial(n):
    hasil = 1

    for i in range(1, n + 1):
        hasil *= i

    return hasil

angka = int(input("Masukkan angka: "))

print("Faktorial =", faktorial(angka))
```

> Kesimpulan

Program menggunakan fungsi untuk menghitung faktorial suatu bilangan. Proses perhitungan dilakukan dengan mengalikan angka secara berurutan mulai dari 1 hingga angka yang dimasukkan. Hasil yang diperoleh sesuai dengan konsep matematis faktorial.

> Kesimpulan

Fungsi dapat digunakan untuk mengimplementasikan berbagai perhitungan matematis secara sistematis dan terstruktur.

##Tugas 15 - Login Sederhana dengan Perulangan

```python
username_benar = "admin"
password_benar = "123"

for i in range(3):
    username = input("Username: ")
    password = input("Password: ")

    if username == username_benar and password == password_benar:
        print("Login berhasil")
        break
else:
    print("Akses ditolak")
```

> Analisis

Program melakukan proses autentikasi dengan membandingkan username dan password yang dimasukkan pengguna dengan data yang telah ditentukan. Pengguna diberikan maksimal tiga kali kesempatan untuk melakukan login. Jika data yang dimasukkan benar, akses diberikan, sedangkan jika seluruh percobaan gagal maka akses ditolak. Hasil percobaan menunjukkan bahwa kombinasi percabangan dan perulangan dapat digunakan untuk mengontrol keamanan akses sistem.

> Kesimpulan

Percabangan dan perulangan dapat diterapkan dalam pembuatan sistem login sederhana untuk mengatur proses autentikasi pengguna secara lebih terkontrol dan aman.
