##DISCUSSION

Soal 1: Apa yang dimaksud dengan algoritma?

=> Jawab

Algoritma adalah sekumpulan langkah-langkah logis dan sistematis yang disusun untuk menyelesaikan suatu masalah tertentu. Setiap langkah dalam algoritma harus jelas, terbatas (ada titik akhirnya), dan menghasilkan output yang diharapkan dari input yang diberikan.

Soal 2: Mengapa algoritma penting dalam pemrograman?

=> Jawab

Algoritma penting karena menjadi blueprint atau kerangka berpikir sebelum kode ditulis. Tanpa algoritma yang baik, program yang dibuat cenderung tidak efisien, sulit di-debug, dan susah dikembangkan.
Selain itu, algoritma yang tepat berpengaruh langsung pada performa program, misalnya dalam hal kecepatan eksekusi dan penggunaan memori.

Soal 3: Apa perbedaan algoritma dan program

= Jawab

> Algoritma

a. Bentuk: Logika/langkah bisa berupa tulisan atau flowchart

b. Sifat: Independen dari bahasa pemrograman

c. Tujuan: Merancang solusi

> Program

a. Bentuk: kode yang ditulis dalam bahasa pemrograman

b. Sifat: terikat pada bahasa tertentu (Python, Java, dll)

c. Tujuan: mengeksekusi solusi di komputer

Soal 4: Mengapa Python cocok untuk pemula?

=> Jawab

a. Syntax yang bersih dan mudah dibaca: Karena Python dirancang mirip dengan bahasa Inggris sehari-hari, sehingga logika program lebih mudah dipahami tanpa terlalu banyak simbol atau aturan yang rumit

b. Tidak perlu deklarasi tipe data secara eksplisit: Python itu bersifat dynamically typed sehingga pemula bisa fokus pada logika, bukan sintaks

c. Serbaguna: Python bisa dipakai untuk web, data science, AI, otomasi, dan lainnya, sehingga ilmu yang dipelajari sejak awal tetap relevan ke depannya
