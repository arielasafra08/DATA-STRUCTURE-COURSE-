##Dasar Teori
> Algoritma

Algoritma adalah serangkaian langkah-langkah logis, terstruktur, dan sistematis yang digunakan untuk menyelesaikan suatu masalah. Dalam ilmu komputer, algoritma menjadi dasar dalam pembuatan program karena berfungsi sebagai panduan untuk menentukan proses yang harus dilakukan komputer dalam menghasilkan suatu keluaran (output).

> Pemrograman Python

Python merupakan bahasa pemrograman tingkat tinggi yang memiliki sintaks sederhana dan mudah dipahami. Python banyak digunakan dalam pengembangan perangkat lunak, analisis data, kecerdasan buatan, machine learning, dan berbagai bidang teknologi lainnya. Karena kemudahan sintaksnya, Python sangat cocok digunakan oleh pemula untuk mempelajari konsep dasar pemrograman.

> Struktur Dasar Pemrograman

Dalam pemrograman terdapat beberapa struktur dasar yang digunakan untuk membangun logika program, yaitu:

- input dan output: digunakan untuk menerima data dari pengguna, sedangkan output digunakan untuk menampilkan hasil pengolahan data.
- Percabangan (Selection): Percabangan digunakan untuk mengambil keputusan berdasarkan kondisi tertentu. Struktur yang umum digunakan adalah if, elif, dan else.
- Perulangan (Looping): Perulangan digunakan untuk menjalankan suatu perintah secara berulang hingga kondisi tertentu terpenuhi.

> Variabel dan Tipe Data

Variabel merupakan tempat penyimpanan data dalam program. Python menyediakan berbagai tipe data seperti integer (int), bilangan desimal (float), teks (string), dan logika (boolean).

##Tugas 1 - Menenrukan Bilangan Positif, Negatif, atau Nol

> Algoritma

a. Input sebuah bilangan.

b. Jika bilangan > 0, tampilkan "Positif".

c. Jika bilangan < 0, tampilkan "Negatif".

d. Jika bilangan = 0, tampilkan "Nol".

```python
bilangan = int(input("Masukkan bilangan: "))

if bilangan > 0:
    print("Bilangan Positif")
elif bilangan < 0:
    print("Bilangan Negatif")
else:
    print("Bilangan Nol")
```

> Analisis

##Tugas 2 - Menampilkan Bilangan Genap dari 1 sampai 20

> Algoritma

a. Mulai dari angka 1.

b. Periksa setiap angka sampai 20.

c. Jika angka habis dibagi 2, tampilkan angka tersebut.

```python
for i in range(1, 21):
    if i % 2 == 0:
        print(i)
```

> Analisis

Program menggunakan perulangan for untuk menampilkan angka dari 1 sampai 20. Operator modulus (%) digunakan untuk mengecek apakah suatu bilangan habis dibagi 2. Jika sisa pembagian adalah 0, maka bilangan tersebut termasuk bilangan genap.

##Tugas 4 - Menentukan Bilangan Terbesar dari 3 Bilangan

> Algoritma

a. Input tiga bilangan.

b. Bandingkan ketiga bilangan.

c. Tentukan bilangan terbesar.

d. Tampilkan hasil.

```python
a = int(input("Masukkan bilangan pertama: "))
b = int(input("Masukkan bilangan kedua: "))
c = int(input("Masukkan bilangan ketiga: "))

terbesar = a

if b > terbesar:
    terbesar = b

if c > terbesar:
    terbesar = c

print("Bilangan terbesar adalah:", terbesar)
```

> Analisis

Program membandingkan tiga bilangan yang diinput pengguna. Variabel terbesar digunakan untuk menyimpan nilai terbesar sementara. Setelah semua perbandingan dilakukan, program menampilkan bilangan terbesar.

##Tugas 5 - Menghitung Rata-rata dan Kategori Nilai

> Algoritma

a. Input 5 nilai.

b. Hitung rata-rata.

c. Jika rata-rata ≥ 80 → Baik.

d. Jika rata-rata 60–79 → Cukup.

e. Jika rata-rata < 60 → Kurang.

f. Tampilkan hasil.

```python
n1 = float(input("Nilai 1: "))
n2 = float(input("Nilai 2: "))
n3 = float(input("Nilai 3: "))
n4 = float(input("Nilai 4: "))
n5 = float(input("Nilai 5: "))

rata = (n1 + n2 + n3 + n4 + n5) / 5

if rata >= 80:
    kategori = "Baik"
elif rata >= 60:
    kategori = "Cukup"
else:
    kategori = "Kurang"

print("Rata-rata =", rata)
print("Kategori =", kategori)
```

> Analisis

Program menghitung rata-rata dari lima nilai yang dimasukkan pengguna. Setelah rata-rata diperoleh, program menentukan kategori berdasarkan rentang nilai yang telah ditentukan.

##Tugas 6 - Menghitung Total Bayar Setelah Diskon

> Algoritma

a. Input total belanja.

b. Jika total belanja ≥ 100.000, berikan diskon 10%.

c. Jika tidak, diskon 0%.

d. Hitung total pembayaran.

e. Tampilkan hasil.

```python
total_belanja = float(input("Masukkan total belanja: "))

if total_belanja >= 100000:
    diskon = total_belanja * 0.10
else:
    diskon = 0

total_bayar = total_belanja - diskon

print("Diskon =", diskon)
print("Total bayar =", total_bayar)
```

> Analisis

Program menggunakan percabangan untuk menentukan apakah pembeli berhak mendapatkan diskon. Jika total belanja mencapai Rp100.000 atau lebih, maka diberikan diskon sebesar 10%. Total pembayaran diperoleh dari total belanja dikurangi diskon.

##Tugas 7 - Menampilkan Tabel Perkalian

> Algoritma

a. Input sebuah angka.

b. Ulangi perkalian dari 1 sampai 10.

c. Tampilkan hasil perkalian.

```python
angka = int(input("Masukkan angka: "))

for i in range(1, 11):
    print(f"{angka} x {i} = {angka * i}")
```

> Analisis

Program menggunakan perulangan for untuk mengalikan angka yang diinput pengguna dengan angka 1 sampai 10. Hasil setiap perkalian ditampilkan dalam bentuk tabel.

##Tugas 8 - Menghitung Faktorial Bilangan

> Algoritma

a. Input sebuah bilangan.

b. Inisialisasi hasil = 1.

c. Kalikan hasil dengan setiap angka dari 1 sampai bilangan tersebut.

d. Tampilkan hasil faktorial.

```python
n = int(input("Masukkan bilangan: "))

faktorial = 1

for i in range(1, n + 1):
    faktorial *= i

print("Faktorial =", faktorial)
```

> Analisis

Faktorial adalah hasil perkalian bilangan dari 1 hingga bilangan tersebut. Misalnya 5! = 1 × 2 × 3 × 4 × 5 = 120. Program menggunakan perulangan untuk melakukan proses perkalian secara berulang.

##Tugas 9 - Menentukan Bilangan Prima

> Algoritma

a. Input sebuah bilangan.

b. Hitung jumlah pembagi bilangan tersebut.

c. Jika hanya memiliki 2 pembagi (1 dan dirinya sendiri), maka prima.

d. Jika lebih dari 2 pembagi, maka bukan prima.

```python
bilangan = int(input("Masukkan bilangan: "))

if bilangan > 1:
    prima = True

    for i in range(2, bilangan):
        if bilangan % i == 0:
            prima = False
            break

    if prima:
        print("Bilangan Prima")
    else:
        print("Bukan Bilangan Prima")
else:
    print("Bukan Bilangan Prima")
```

> Analisis

Bilangan prima adalah bilangan yang hanya memiliki dua faktor, yaitu 1 dan dirinya sendiri. Program memeriksa apakah ada bilangan lain yang dapat membagi habis bilangan tersebut. Jika ada, maka bilangan tersebut bukan prima.

##Tugas 10 - Program Login dengan Maksimal 3 Percobaan

> Algoritma

a. Tentukan username = "admin" dan password = "123".

b. Berikan kesempatan login maksimal 3 kali.

c. Jika username dan password benar, tampilkan "Login Berhasil".

d. Jika salah sebanyak 3 kali, tampilkan "Akses Ditolak".

```python
username_benar = "admin"
password_benar = "123"

for percobaan in range(3):
    username = input("Username: ")
    password = input("Password: ")

    if username == username_benar and password == password_benar:
        print("Login Berhasil")
        break
    else:
        print("Username atau Password salah")

else:
    print("Akses Ditolak")
```

> Analisis

Program menggunakan perulangan sebanyak tiga kali sebagai batas percobaan login. Jika pengguna memasukkan username dan password yang sesuai, program akan menampilkan pesan "Login Berhasil" dan menghentikan perulangan. Jika sampai tiga kali percobaan masih salah, program menampilkan pesan "Akses Ditolak".
