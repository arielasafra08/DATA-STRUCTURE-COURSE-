##Dasar Teori
> Variabel

Wadah penyimpan data yang nilainya bisa berubah-ubah. Ibaratnya seperti kotak berlabel, labelnya nama variabel, isinya nilainya.
> Tipe Data

Jenis/kategori data yang disimpan. Komputer perlu tau ini supaya tahu cara memprosesnya. Terdiri dari int (bilangan bulat), float (bilangan desimal), string (teks), boolean (benar/salah)
> Percabangan

Struktur yang membuat program bisa memilih jalur berdasarkan kondisi. Kalau kondisi terpenuhi → jalur A, kalau tidak → jalur B.
Kata kuncinya: if, elif, else
> Perulangan

Perintah untuk mengeksekusi blok kode berulang kali tanpa menulis ulang. Cocok untuk tugas yang repetitif.
Kata kuncinya: for (sudah tahu berapa kali), while (selama kondisi terpenuhi)
> List/Array

Wadah yang bisa menyimpan banyak data sekaligus dalam satu variabel, tersusun berurutan dan bisa diakses lewat indeks (dimulai dari 0)

##Studi Kasus 6: Analisis Nilai Mahasiswa (Gabungan)

```python
def hitung_rata(data):
    return sum(data) / len(data)

nilai = []

for i in range(5):
    n = int(input(f"Masukkan nilai ke-{i+1}: "))
    nilai.append(n)

rata = hitung_rata(nilai)

if rata >= 80:
    kategori = "Baik"
elif rata >= 60:
    kategori = "Cukup"
else:
    kategori = "Kurang"

print("\nRata-rata =", rata)
print("Kategori =", kategori)
```

##Studi Kasus 7: Sistem Kasir Sederhana

```python
harga1 = int(input("Masukkan harga barang 1: "))
harga2 = int(input("Masukkan harga barang 2: "))
harga3 = int(input("Masukkan harga barang 3: "))

total = harga1 + harga2 + harga3

if total >= 100000:
    diskon = total * 0.10
else:
    diskon = 0

total_bayar = total - diskon

print("\nTotal Belanja = Rp", total)
print("Diskon = Rp", diskon)
print("Total Bayar = Rp", total_bayar)
```

##Studi Kasus 9: Pencarian Nilai Tertinggi

```python
jumlah = int(input("Berapa banyak nilai yang akan diinput? "))

nilai = []

for i in range(jumlah):
    n = int(input(f"Masukkan nilai ke-{i+1}: "))
    nilai.append(n)

tertinggi = nilai[0]

for n in nilai:
    if n > tertinggi:
        tertinggi = n

print("\nDaftar Nilai:", nilai)
print("Nilai Tertinggi:", tertinggi)
```

##Studi Kasus 10: Sistem Login Sederhana

```python
username_benar = "admin"
password_benar = "123"

percobaan = 0

while percobaan < 3:
    username = input("Masukkan Username: ")
    password = input("Masukkan Password: ")

    if username == username_benar and password == password_benar:
        print("Login Berhasil")
        break
    else:
        percobaan += 1
        print("Username atau Password Salah")

if percobaan == 3:
    print("Akses Ditolak")
```
