##Praktikum 3 - Bagian Queue menggunakan Deque

```python
from collections import deque

queue = deque()

queue.append("A")
queue.append("B")
queue.append("C")

print(queue)
```

##Praktikum 4 - Simulasi Antrian Bank

```python
from collections import deque

antrian = deque()

# Ambil nomor antrean
antrian.append("Nasabah 1")
antrian.append("Nasabah 2")
antrian.append("Nasabah 3")

print("Antrian saat ini:")
print(antrian)

# Panggil nasabah
dipanggil = antrian.popleft()
print("\nNasabah yang dipanggil:", dipanggil)

# Tampilkan sisa antrean
print("Sisa antrean:")
print(antrian)
```

##Praktikum 5 - Simulasi Printer Queue

```python
from collections import deque

printer = deque()

# Tambah dokumen
printer.append("Laporan.pdf")
printer.append("Tugas.docx")
printer.append("Sertifikat.pdf")

print("Antrian cetak:")
print(printer)

# Cetak dokumen
dokumen = printer.popleft()
print("\nMencetak:", dokumen)

# Lihat antrean tersisa
print("Sisa antrean cetak:")
print(printer)
```

##Praktikum 6 - Simulasi Deque

```python
from collections import deque

data = deque()

# append()
data.append(10)
data.append(20)

# appendleft()
data.appendleft(5)

print("Setelah append dan appendleft:")
print(data)

# pop()
hapus_belakang = data.pop()
print("\nHapus dari belakang:", hapus_belakang)
print(data)

# popleft()
hapus_depan = data.popleft()
print("\nHapus dari depan:", hapus_depan)
print(data)
```

##DISCUSSION

Soal 1: Apa yang dimaksud dengan Queue?

=> Jawab:
Queue adalah struktur data yang bekerja seperti antrean, di mana data yang masuk lebih dulu akan keluar lebih dulu (FIFO).

Soal 2: Mengapa Queue menggunakan prinsip FIFO?

=> Jawab: Karena Queue dirancang untuk melayani data sesuai urutan kedatangannya sehingga proses menjadi lebih adil dan teratur.

Soal 3: Apa perbedaan Enqueue dan Dequeue?

=> Jawab: Enqueue itu menambahkan data ke belakang antrean, sedangkan dequeue itu menghapus data dari depan antrean.

Soal 4: Apa perbedaan Queue dan Stack?

=> Jawab: Queue menggunakan prinsip FIFO (First In First Out), sedangkan stack menggunakan prinsip LIFO (Last In First Out).

Soal 5: Apa kelebihan Deque dibanding Queue?

=> Jawab: Deque lebih fleksibel karena penambahan dan penghapusan data dapat dilakukan dari depan maupun belakang.

Soal 6: Mengapa deque lebih efisien dibanding List untuk Queue?

=> Jawab: Karena operasi penambahan dan penghapusan di deque memiliki kompleksitas O(1), sedangkan pada list penghapusan di depan memerlukan pergeseran elemen sehingga lebih lambat.

Soal 7: Sebutkan penerapan Queue dalam sistem operasi.

=> Jawab: Queue digunakan pada process scheduling, manajemen proses CPU, printer queue, dan pengaturan tugas yang menunggu untuk diproses.

Soal 8: Bagaimana Queue digunakan pada layanan perbankan?

=> Jawab: Nasabah yang mengambil nomor antrean lebih dahulu akan dilayani lebih dahulu sesuai urutan kedatangan.

Soal 9: Apa yang dimaksud dengan Process Scheduling?

=> Jawab: Process Scheduling adalah mekanisme sistem operasi untuk mengatur urutan proses yang akan menggunakan CPU agar berjalan secara efisien.

Soal 10: Mengapa Queue penting dalam sistem informasi modern?

=> Jawab: Karena Queue membantu mengatur proses secara terurut, menghindari konflik data, meningkatkan efisiensi sistem, dan memastikan setiap tugas diproses sesuai urutan.
