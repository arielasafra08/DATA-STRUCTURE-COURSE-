##Praktikum 3 - Transversal Linked List

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

# Membuat node
node1 = Node(10)
node2 = Node(20)
node3 = Node(30)
node4 = Node(40)

# Menghubungkan node
node1.next = node2
node2.next = node3
node3.next = node4

# Traversal
current = node1
while current:
    print(current.data, end=" -> ")
    current = current.next

print("NULL")
```

##Praktikum 4 - Insert Node

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

# Linked List awal
node1 = Node(10)
node2 = Node(20)
node3 = Node(30)

node1.next = node2
node2.next = node3

# Insert awal
new_awal = Node(5)
new_awal.next = node1
head = new_awal

# Insert akhir
new_akhir = Node(40)

current = head
while current.next:
    current = current.next

current.next = new_akhir

# Tampilkan Linked List
current = head
while current:
    print(current.data, end=" -> ")
    current = current.next

print("NULL")
```

##Praktikum 5 - Delete Node

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

# Linked List
node1 = Node(10)
node2 = Node(20)
node3 = Node(30)

node1.next = node2
node2.next = node3

head = node1

# Hapus node bernilai 20
current = head

while current.next:
    if current.next.data == 20:
        current.next = current.next.next
        break
    current = current.next

# Tampilkan hasil
current = head
while current:
    print(current.data, end=" -> ")
    current = current.next

print("NULL")
```

##Praktikum 6 - Search Data

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

# Linked List
node1 = Node(10)
node2 = Node(20)
node3 = Node(30)

node1.next = node2
node2.next = node3

cari = 20
current = node1
ditemukan = False

while current:
    if current.data == cari:
        ditemukan = True
        break
    current = current.next

if ditemukan:
    print("Data ditemukan")
else:
    print("Data tidak ditemukan")
```

##Praktikum 7 - Doubly Linked List

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.prev = None
        self.next = None

# Membuat node
node1 = Node("A")
node2 = Node("B")
node3 = Node("C")

# Menghubungkan node
node1.next = node2
node2.prev = node1

node2.next = node3
node3.prev = node2

# Traversal maju
current = node1

while current:
    print(current.data, end=" <-> ")
    current = current.next

print("NULL")
```

##Praktikum 8 - Circular Linked List

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

# Membuat node
node1 = Node("A")
node2 = Node("B")
node3 = Node("C")

# Menghubungkan secara melingkar
node1.next = node2
node2.next = node3
node3.next = node1

# Menampilkan 6 node agar terlihat melingkar
current = node1

for i in range(6):
    print(current.data, end=" -> ")
    current = current.next

print("...")
```

##DISCUSSION

Soal 1: Apa yang dimaksud dengan Linked List?

Linked List adalah struktur data yang terdiri dari beberapa node yang saling terhubung melalui pointer atau referensi.

Soal 2: Mengapa Linked List disebut struktur data dinamis?

Karena ukuran Linked List dapat bertambah atau berkurang sesuai kebutuhan tanpa harus menentukan ukuran sejak awal.

Soal 3: Apa perbedaan Array dan Linked List?


Array menyimpan data pada memori yang berurutan dan akses data lebih cepat.
Linked List menyimpan data secara tidak berurutan dan setiap node saling terhubung menggunakan pointer.

Soal 4: Apa fungsi pointer pada Linked List?

Pointer berfungsi untuk menyimpan alamat atau referensi node berikutnya sehingga node-node dapat saling terhubung.

Soal 5: Apa perbedaan Singly dan Doubly Linked List?

Singly Linked List hanya memiliki pointer ke node berikutnya.
Doubly Linked List memiliki dua pointer, yaitu ke node sebelumnya dan node berikutnya.

Soal 6: Apa keuntungan Circular Linked List?

Apa keuntungan Circular Linked List?
Node terakhir terhubung kembali ke node pertama sehingga proses perulangan data dapat dilakukan tanpa harus kembali ke awal secara manual.
Node terakhir terhubung kembali ke node pertama sehingga proses perulangan data dapat dilakukan tanpa harus kembali ke awal secara manual.

Soal 7: Mengapa akses data pada Linked List lebih lambat dibanding Array?

Karena untuk mencari suatu data, Linked List harus menelusuri node satu per satu dari awal hingga data ditemukan.

Soal 8: Pada kondisi apa Linked List lebih baik digunakan dibanding Array?

Saat data sering mengalami penambahan atau penghapusan, karena Linked List lebih fleksibel dan tidak memerlukan pergeseran elemen seperti Array.

- Playlist musik

- Undo dan redo pada editor teks
- Manajemen memori sistem operasi
- Sistem reservasi dan navigasi

Soal 10: Mengapa Linked List menjadi dasar banyak struktur data lain?

Karena Linked List mudah dikembangkan untuk membentuk struktur data lain seperti Stack, Queue, Hash Table, dan Graph sehingga lebih fleksibel dalam berbagai kebutuhan pemrograman.
