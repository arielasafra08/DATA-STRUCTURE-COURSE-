##Dasar Teori

Stack adalah struktur data linear yang bekerja menggunakan prinsip LIFO (Last In First Out), yaitu data yang terakhir masuk akan menjadi data pertama yang keluar. Operasi utama pada Stack meliputi Push (menambah data), Pop (menghapus data teratas), Peek (melihat data teratas), IsEmpty (mengecek apakah Stack kosong), dan Size (menghitung jumlah elemen). Dalam Python, Stack dapat diimplementasikan menggunakan list dengan fungsi append() dan pop(). Struktur data ini banyak digunakan pada fitur Undo-Redo, Browser History, Call Stack, dan validasi tanda kurung.

##Tugas 4 - Menu Stack Interaktif

```python
stack = []

while True:
    print("\n=== MENU STACK ===")
    print("1. Push")
    print("2. Pop")
    print("3. Peek")
    print("4. Tampilkan Stack")
    print("5. Keluar")

    pilihan = input("Pilih menu: ")

    if pilihan == "1":
        data = input("Masukkan data: ")
        stack.append(data)

    elif pilihan == "2":
        if stack:
            print("Data keluar:", stack.pop())
        else:
            print("Stack kosong")

    elif pilihan == "3":
        if stack:
            print("Elemen teratas:", stack[-1])
        else:
            print("Stack kosong")

    elif pilihan == "4":
        print("Isi Stack:", stack)

    elif pilihan == "5":
        break

    else:
        print("Pilihan tidak valid")
```

> Langkah Kerja

1. Membuat menu menggunakan perulangan.
2. Menambahkan fitur Push, Pop, Peek, dan Display.
3. Program berhenti saat memilih menu keluar.

> Analisis

Program memungkinkan pengguna mengelola Stack secara dinamis melalui menu interaktif.

> Kesimpulan

##Tugas 7 - Validasi Tanda Kurung

```python
def cek_kurung(ekspresi):
    stack = []

    for karakter in ekspresi:
        if karakter == "(":
            stack.append(karakter)

        elif karakter == ")":
            if not stack:
                return "Tidak Valid"
            stack.pop()

    if not stack:
        return "Valid"
    else:
        return "Tidak Valid"

data = input("Masukkan ekspresi: ")
print(cek_kurung(data))
```

> Langkah Kerja

1. Membaca ekspresi.
2. Menyimpan tanda kurung buka ke Stack.
3. Menghapusnya saat menemukan kurung tutup.
4. Mengecek apakah Stack kosong.

> Analisis

Jika semua pasangan kurung cocok maka Stack kosong di akhir proses sehingga ekspresi valid.

> Kesimpulan

Stack sangat efektif digunakan untuk pengecekan pasangan tanda kurung.

##Tugas 8 - Simulasi Browser History

```python
history = []

while True:
    print("\n1. Kunjungi Halaman")
    print("2. Back")
    print("3. Lihat History")
    print("4. Keluar")

    pilih = input("Pilih: ")

    if pilih == "1":
        halaman = input("Nama halaman: ")
        history.append(halaman)

    elif pilih == "2":
        if history:
            print("Kembali dari:", history.pop())
        else:
            print("History kosong")

    elif pilih == "3":
        print("Riwayat:", history)

    elif pilih == "4":
        break
```

> Langkah Kerja

1. Membuat Stack history.
2. Menambahkan halaman yang dikunjungi.
3. Menggunakan Pop saat tombol Back ditekan.
4. Menampilkan riwayat.

> Analisis

Halaman terakhir yang dikunjungi akan menjadi halaman pertama yang ditinggalkan saat Back ditekan.

> Kesimpulan

Fitur Browser History merupakan penerapan langsung konsep Stack.

##Tugas 9 - Sistem Undo-Redo Sederhana

```python
undo_stack = []
redo_stack = []

while True:
    print("\n1. Tulis")
    print("2. Undo")
    print("3. Redo")
    print("4. Keluar")

    pilih = input("Pilih: ")

    if pilih == "1":
        teks = input("Masukkan teks: ")
        undo_stack.append(teks)
        redo_stack.clear()

    elif pilih == "2":
        if undo_stack:
            data = undo_stack.pop()
            redo_stack.append(data)
            print("Undo:", data)
        else:
            print("Tidak ada data")

    elif pilih == "3":
        if redo_stack:
            data = redo_stack.pop()
            undo_stack.append(data)
            print("Redo:", data)
        else:
            print("Tidak ada data")

    elif pilih == "4":
        break
```

> Langkah Kerja

1. Membuat Stack Undo dan Redo.
2. Menyimpan perubahan pada Undo Stack.
3. Memindahkan data ke Redo Stack saat Undo.
4. Mengembalikan data saat Redo.

> Analisis

Program menggunakan dua Stack agar proses Undo dan Redo berjalan dengan benar.

> Kesimpulan

Sistem Undo-Redo dapat diimplementasikan menggunakan dua buah Stack.

##Tugas 10 - Analisis Kompleksitas

| Operasi | Kompleksitas |
| ------- | ------------ |
| Push    | O(1)         |
| Pop     | O(1)         |
| Peek    | O(1)         |
| IsEmpty | O(1)         |
| Size    | O(1)         |

> Analisis Browser History

- Kunjungi Halaman → Push → O(1)
- Back → Pop → O(1)
- Lihat Halaman Terakhir → Peek → O(1)

> Analisis Undo-Redo

- Menyimpan Perubahan → Push → O(1)
- Undo → Pop + Push → O(1)
- Redo → Pop + Push → O(1)

> Kesimpulan

Seluruh operasi dasar Stack memiliki kompleksitas O(1) karena hanya mengakses elemen pada bagian atas Stack sehingga prosesnya sangat cepat dan efisien.
