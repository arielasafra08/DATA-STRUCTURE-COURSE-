##Dasar Teori

Queue adalah struktur data linear yang bekerja berdasarkan prinsip FIFO (First In First Out), yaitu data yang pertama masuk akan menjadi data pertama yang keluar. Queue memiliki dua bagian utama, yaitu Front sebagai posisi elemen pertama dan Rear sebagai posisi elemen terakhir. Operasi dasar pada Queue meliputi Enqueue untuk menambahkan data ke belakang antrian dan Dequeue untuk menghapus data dari depan antrian.

Selain Queue, terdapat Deque (Double Ended Queue) yang merupakan pengembangan dari Queue. Deque memungkinkan penambahan dan penghapusan data dilakukan dari kedua sisi, baik di bagian depan maupun belakang. Dalam Python, Deque dapat diimplementasikan menggunakan modul collections.deque yang lebih efisien dibandingkan list untuk operasi antrian.

Queue dan Deque banyak digunakan dalam kehidupan sehari-hari maupun sistem komputer, seperti antrian nasabah di bank, antrian pasien di rumah sakit, printer queue, process scheduling pada sistem operasi, call center, serta pengelolaan playlist musik. Penggunaan Queue dan Deque membantu proses pengolahan data menjadi lebih teratur, efisien, dan sesuai urutan yang diinginkan.

##Tugas 2 - Operasi Enqueue dan Deque

```python
queue = []

queue.append("Kit")
queue.append("Dorothy")
queue.append("Lili")

print("Queue Awal:", queue)

layani = queue.pop(0)

print("Dilayani:", layani)
print("Queue Sekarang:", queue)
```

> Langkah Kerja

1. Membuat Queue.
2. Menambahkan data menggunakan Enqueue (append()).
3. Menghapus data pertama menggunakan Dequeue (pop(0)).
4. Menampilkan hasil.

> Analisis

Saat Dequeue dilakukan, elemen pertama yaitu Kit keluar lebih dulu karena Queue menggunakan FIFO.

> Kesimpulan

Enqueue digunakan untuk menambah data di belakang antrian, sedangkan Dequeue digunakan untuk menghapus data di depan antrian.

##Tugas 4 - Menu Queue Interaktif

```python
queue = []

while True:
    print("\n=== MENU QUEUE ===")
    print("1. Enqueue")
    print("2. Dequeue")
    print("3. Lihat Queue")
    print("4. Keluar")

    pilih = input("Pilih menu: ")

    if pilih == "1":
        data = input("Masukkan data: ")
        queue.append(data)

    elif pilih == "2":
        if queue:
            print("Dilayani:", queue.pop(0))
        else:
            print("Queue kosong")

    elif pilih == "3":
        print("Isi Queue:", queue)

    elif pilih == "4":
        break
```

##Tugas 6 - Implementasi Queue menggunakan Deque

```python
from collections import deque

queue = deque()

queue.append("Kit")
queue.append("Branch")
queue.append("Lili")

print(queue)

queue.popleft()

print(queue)
```

> Langkah Kerja

1. Mengimpor modul deque.
2. Membuat Queue menggunakan deque.
3. Menambahkan data dengan append().
4. Menghapus data depan dengan popleft().

> Analisis

Deque lebih efisien dibanding list karena penghapusan data depan tidak memerlukan pergeseran elemen.

> Kesimpulan

Deque merupakan implementasi Queue yang lebih cepat dan efisien.

##Tugas 7 - Perbandingan Queue List dan Deque

```python
from collections import deque

# List
queue_list = ["A", "B", "C"]
queue_list.pop(0)

# Deque
queue_deque = deque(["A", "B", "C"])
queue_deque.popleft()
```

> Langkah Kerja

1. Membuat Queue menggunakan List.
2. Membuat Queue menggunakan Deque.
3. Melakukan operasi penghapusan data depan.
4. Membandingkan kinerjanya.

> Analisis

| Operasi         | List         | Deque       |
| --------------- | ------------ | ----------- |
| Tambah Belakang | O(1)         | O(1)        |
| Hapus Depan     | O(n)         | O(1)        |
| Efisiensi       | Lebih Lambat | Lebih Cepat |


> Kesimpulan

Deque lebih direkomendasikan untuk implementasi Queue karena lebih efisien.

##Tugas 11 - Implementasi Deque untuk Playlist Musik

```python
from collections import deque

playlist = deque()

playlist.append("Mrs. Magic Song")
playlist.append("Golden Brown Song")
playlist.appendleft("Luminary Song")

print("Playlist:", playlist)

playlist.pop()
print("Setelah pop:", playlist)

playlist.popleft()
print("Setelah popleft:", playlist)
```

> Langkah Kerja

1. Membuat playlist menggunakan deque.
2. Menambah lagu dari depan dan belakang.
3. Menghapus lagu dari depan dan belakang.
4. Menampilkan hasil.

> Analisis

Deque memberikan fleksibilitas untuk mengelola playlist dari kedua sisi.

> Kesimpulan

Deque lebih fleksibel dibanding Queue biasa karena dapat melakukan operasi pada kedua ujung.

##Tugas 12 - Analisis Kompleksitas

| Operasi             | Kompleksitas |
| ------------------- | ------------ |
| Enqueue (`append`)  | O(1)         |
| Dequeue (`popleft`) | O(1)         |
| Front               | O(1)         |
| Rear                | O(1)         |
| Appendleft          | O(1)         |
| Pop                 | O(1)         |
| Popleft             | O(1)         |


> Analisis

- Sistem Rumah Sakit menggunakan append() dan popleft() → O(1)
- Printer Queue menggunakan append() dan popleft() → O(1)
- Call Center menggunakan append() dan popleft() → O(1)
- Playlist Musik menggunakan append(), appendleft(), pop(), dan popleft() → O(1)

Implementasi Queue dan Deque menggunakan collections.deque sangat efisien karena hampir seluruh operasinya memiliki kompleksitas O(1) sehingga cocok digunakan untuk pengolahan data secara real-time dan skala besar.
