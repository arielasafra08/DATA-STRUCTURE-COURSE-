##Bagian Simulasi Browser History

```python
history = []

while True:
    print("\n=== Browser History ===")
    print("1. Kunjungi Halaman")
    print("2. Back")
    print("3. Tampilkan Riwayat")
    print("4. Keluar")

    pilihan = input("Pilih menu (1-4): ")

    if pilihan == "1":
        halaman = input("Masukkan nama halaman: ")
        history.append(halaman)
        print(f"Berhasil membuka {halaman}")

    elif pilihan == "2":
        if len(history) > 0:
            keluar = history.pop()
            print(f"Kembali dari halaman {keluar}")
        else:
            print("Riwayat kosong")

    elif pilihan == "3":
        if len(history) > 0:
            print("\nRiwayat Halaman:")
            for i in history:
                print("-", i)
        else:
            print("Belum ada riwayat")

    elif pilihan == "4":
        print("Program selesai")
        break

    else:
        print("Pilihan tidak valid")
```

##Bagian Validasi Kurung

```python
ekspresi = input("Masukkan ekspresi: ")

stack = []
valid = True

for karakter in ekspresi:
    if karakter == "(":
        stack.append(karakter)

    elif karakter == ")":
        if len(stack) == 0:
            valid = False
            break
        stack.pop()

if len(stack) != 0:
    valid = False

if valid:
    print("Valid")
else:
    print("Tidak Valid")
```

##DISCUSSION

Soal 1: Apa yang dimaksud dengan stack

=> Jawab: Stack adalah struktur data linear yang menggunakan prinsip LIFO (Last In First Out), yaitu data yang terakhir masuk akan menjadi data pertama yang keluar.

Soal 2: Mengapa stack menggunakan prinsip LIFO

=> Jawab: Karena seluruh operasi dilakukan pada satu sisi, yaitu bagian atas (top), sehingga elemen terakhir yang masuk akan diakses terlebih dahulu.

Soal 3: Apa perbedaan push dan pop

=> Jawab: Push menambahkan data ke bagian atas Stack, sedangkan pop menghapus dan mengambil data dari bagian atas Stack.

Soal 4: Mengapa operasi stack memiliki kompleksitas O(1)

=> Jawab: Karena operasi hanya dilakukan pada elemen teratas tanpa perlu menelusuri seluruh isi Stack.

Soal 5: Apa fungsi call stack

=> Jawab: Call Stack berfungsi untuk menyimpan informasi pemanggilan fungsi saat program berjalan sehingga program dapat kembali ke fungsi sebelumnya setelah selesai dieksekusi.

Soal 6: Apa hubungan stack dengan rekursi

=> Jawab: Setiap pemanggilan fungsi rekursif akan disimpan ke dalam Call Stack. Saat fungsi selesai, data akan dikeluarkan dari Stack secara berurutan.

Soal 7: Apa yang menyebabkan stack overflow

=> Jawab: Stack Overflow terjadi karena terlalu banyak pemanggilan fungsi atau rekursi sehingga kapasitas memori Stack habis.

Soal 8: Mengapa fitur undo menggunakan stack

=> Jawab: Karena perubahan terakhir yang dilakukan pengguna harus dibatalkan terlebih dahulu, sesuai prinsip LIFO.

Soal 9: Apa kelebihan dan kekurangan stack

=> Jawab:

> Kelebihan:
- Implementasi sederhana.
- Operasi Push dan Pop cepat (O(1)).
- Cocok untuk Undo, Back, dan rekursi.

> Kekurangan:

- Akses data terbatas hanya pada elemen teratas.
- Tidak dapat mengakses elemen tengah secara langsung.
- Kurang fleksibel untuk pengolahan data yang kompleks.
