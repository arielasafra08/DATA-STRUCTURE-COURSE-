##Dasar Teori

Linked List adalah struktur data linear yang terdiri dari sekumpulan node yang saling terhubung melalui pointer atau referensi. Setiap node memiliki dua bagian utama, yaitu data untuk menyimpan nilai dan pointer (next) yang digunakan untuk menunjuk ke node berikutnya. Berbeda dengan array yang menyimpan data secara berurutan di memori, Linked List menyimpan data secara dinamis sehingga lebih fleksibel dalam proses penambahan dan penghapusan data.

Terdapat beberapa jenis Linked List, yaitu Singly Linked List, Doubly Linked List, dan Circular Linked List. Singly Linked List hanya memiliki satu pointer ke node berikutnya, Doubly Linked List memiliki pointer ke node sebelumnya dan berikutnya, sedangkan Circular Linked List menghubungkan node terakhir kembali ke node pertama sehingga membentuk lingkaran.

Operasi dasar pada Linked List meliputi traversal untuk menampilkan seluruh data, insertion untuk menambahkan node baru, deletion untuk menghapus node, dan searching untuk mencari data tertentu. Struktur data ini banyak digunakan dalam berbagai aplikasi seperti playlist musik, browser history, sistem reservasi, editor teks, manajemen memori, serta sebagai dasar implementasi struktur data lain seperti Stack dan Queue.

##Tugas 2 - Membuat Lnkked List Berisi 5 Angka

##Tugas 3 - Traversal Seluruh Node

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

node1 = Node(10)
node2 = Node(20)
node3 = Node(30)
node4 = Node(40)
node5 = Node(50)

node1.next = node2
node2.next = node3
node3.next = node4
node4.next = node5

current = node1

while current:
    print(current.data)
    current = current.next
```

> Langkah Kerja

1. Membuat Linked List.
2. Menentukan node pertama sebagai awal traversal.
3. Menampilkan data satu per satu hingga node terakhir.

> Analisis

Traversal dilakukan dengan mengikuti hubungan antar node dari awal sampai akhir.

> Kesimpulan

Traversal digunakan untuk membaca seluruh data yang terdapat dalam Linked List.

##Tugas 4 - Menambahkan Node di Awal Linked List

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

head = Node(20)
head.next = Node(30)

node_baru = Node(10)
node_baru.next = head
head = node_baru

current = head
while current:
    print(current.data)
    current = current.next
```

> Langkah Kerja

1. Membuat Linked List awal.
2. Membuat node baru dengan nilai 10.
3. Menghubungkan node baru ke head lama.
4. Menjadikan node baru sebagai head.

> Analisis

Penambahan node di awal cukup mengubah referensi head tanpa menggeser data lain.

> Kesimpulan

Insert di awal Linked List lebih efisien dibandingkan Array.

##Tugas 5 - Menambahkan Node di Akhir Linked List

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

head = Node(10)
head.next = Node(20)

node_baru = Node(30)

current = head
while current.next:
    current = current.next

current.next = node_baru

current = head
while current:
    print(current.data)
    current = current.next
```

> Langkah Kerja

1. Membuat Linked List.
2. Menelusuri hingga node terakhir.
3. Menambahkan node baru di akhir.

> Analisis

Program harus mencari node terakhir terlebih dahulu sebelum melakukan penambahan.

> Kesimpulan

Penambahan node di akhir Linked List memerlukan proses traversal terlebih dahulu.

##Tugas 6 - Menghapus Node Tertentu

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

head = Node(10)
node2 = Node(20)
node3 = Node(30)

head.next = node2
node2.next = node3

head.next = node3

current = head
while current:
    print(current.data)
    current = current.next
```

> Langkah Kerja

> Analisis

> Kesimpulan

Penghapusan node dilakukan dengan mengubah referensi antar node.

##Tugas 7 - Mencari Data pada Linked List

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

head = Node(10)
head.next = Node(20)
head.next.next = Node(30)

cari = 20
current = head

while current:
    if current.data == cari:
        print("Data ditemukan")
        break
    current = current.next
else:
    print("Data tidak ditemukan")
```

> Langkah Kerja

1. Membuat Linked List.
2. Menentukan data yang dicari.
3. Melakukan pencarian dari node pertama hingga terakhir.

> Analisis

Setiap node diperiksa satu per satu sampai data ditemukan.

> Kesimpulan

Proses pencarian pada Linked List dilakukan secara berurutan.

##Tugas 8 - Menghitung Jumlah Node

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

head = Node(10)
head.next = Node(20)
head.next.next = Node(30)

jumlah = 0
current = head

while current:
    jumlah += 1
    current = current.next

print("Jumlah node:", jumlah)
```

> Langkah Kerja

1. Membuat Linked List.
2. Menelusuri seluruh node.
3. Menghitung jumlah node yang ada.

> Analisis

Variabel penghitung bertambah setiap kali menemukan node baru.

> Kesimpulan

Jumlah node diperoleh dengan melakukan traversal pada seluruh Linked List.

##Tugas 9 - Implementasi Doubly Linked List

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.prev = None
        self.next = None

node1 = Node("A")
node2 = Node("B")
node3 = Node("C")

node1.next = node2
node2.prev = node1

node2.next = node3
node3.prev = node2

print(node1.data, node2.data, node3.data)
```

> Langkah Kerja

1. Membuat node dengan pointer prev dan next.
2. Menghubungkan node ke depan dan belakang.
3. Menampilkan data.

> Analisis

Setiap node dapat diakses dari dua arah sehingga lebih fleksibel.

> Kesimpulan

Doubly Linked List memungkinkan navigasi maju dan mundur.

##Tugas 10 - Implementasi Circular Linked List

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

node1 = Node("A")
node2 = Node("B")
node3 = Node("C")

node1.next = node2
node2.next = node3
node3.next = node1

print("Circular Linked List berhasil dibuat")
```

> Langkah Kerja

1. Membuat beberapa node.
2. Menghubungkan node terakhir ke node pertama.
3. Membentuk struktur melingkar.

> Analisis

Tidak terdapat node yang menunjuk ke None karena node terakhir kembali ke node pertama.

> Kesimpulan

Circular Linked List cocok digunakan untuk sistem yang berulang secara terus-menerus.

##Tugas 11 - Sistem Playlist Musik menggunakan Linked List

```python
class Lagu:
    def __init__(self, judul):
        self.judul = judul
        self.next = None

lagu1 = Lagu("Mrs. Magic")
lagu2 = Lagu("Golden Brown")
lagu3 = Lagu("Luminary")

lagu1.next = lagu2
lagu2.next = lagu3

current = lagu1

while current:
    print(current.judul)
    current = current.next
```

> Langkah Kerja

1. Membuat node lagu.
2. Menghubungkan setiap lagu.
3. Menampilkan playlist.

> Analisis

Setiap lagu terhubung ke lagu berikutnya sehingga membentuk daftar putar sederhana.

> Kesimpulan

Linked List dapat digunakan untuk mengelola playlist musik secara dinamis.

##Tugas 12 - Browser History menggunakan Doubly Linked List

```python
class Halaman:
    def __init__(self, url):
        self.url = url
        self.prev = None
        self.next = None

google = Halaman("google.com")
youtube = Halaman("youtube.com")
github = Halaman("github.com")

google.next = youtube
youtube.prev = google

youtube.next = github
github.prev = youtube

print("Halaman saat ini:", youtube.url)
print("Back:", youtube.prev.url)
print("Forward:", youtube.next.url)
```

> Langkah Kerja

1. Membuat beberapa halaman web.
2. Menghubungkan halaman menggunakan prev dan next.
3. Menampilkan fitur back dan forward.

> Analisis

Doubly Linked List memungkinkan perpindahan halaman ke depan maupun ke belakang dengan mudah.

> Kesimpulan

Browser history sangat cocok diimplementasikan menggunakan Doubly Linked List karena mendukung navigasi dua arah.
