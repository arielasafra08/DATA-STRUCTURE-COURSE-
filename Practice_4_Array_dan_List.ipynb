##Dasar Teori

> List

Struktur data yang digunakan untuk menyimpan sekumpulan data dengan tipe yang sama dalam satu variabel. Data pada array disusun secara berurutan dan setiap elemen memiliki indeks yang digunakan untuk mengaksesnya. Array memudahkan pengelolaan data dalam jumlah banyak karena data dapat diakses secara langsung berdasarkan posisinya.

> Array

Struktur data linear pada Python yang digunakan untuk menyimpan banyak data dalam satu variabel. Berbeda dengan array, list dapat menyimpan berbagai tipe data dalam satu tempat dan bersifat mutable, yaitu dapat ditambah, diubah, maupun dihapus setelah dibuat. Setiap elemen dalam list memiliki indeks yang dimulai dari 0 sehingga dapat diakses dengan mudah sesuai posisinya.

##PRACTICE

##Bagian Membuat dan Menampilkan List

```python
data = [10, 20, 30, 40]

print(data)
```

##Bagian Menambah dan Menghapus Data

```python
data = [10, 20, 30]

data.append(40)
data.append(20)

print(data)
```

##Bagian Update Data

```python
data = [10, 20, 30]

data[1] = 100

print(data)
```

##Bagian Transversal Data

```python
data = [10, 20, 30, 40]

for i in data:
    print(i)
```

##Bagian Searching Data

```python
data = [10, 20, 30, 40]
cari = 20

for i in data:
    if i == cari:
        print("Data ditemukan")
```

##Bagian Sorting Data

```python
data = [40, 10, 30, 20]

data.sort()

print(data)
```

##STUDY CASE

##Studi Kasus 1 - Data Nilai Mahasiswa
-> Deskripsi: Dosen ingin menyimpan nilai mahasiswa dalam satu struktur data

```python
nilai_mahasiswa = [85, 78, 90, 88, 76]

print("Daftar Nilai Mahasiswa:")
for nilai in nilai_mahasiswa:
    print(nilai)
```

##Studi Kasus 2: Pencarian Data Barang
-> Deskripsi: Sebuah koperasi ingin nama barang tertentu

```python
barang = ["Buku", "Pensil", "Penghapus", "Pulpen", "Penggaris"]

cari = input("Masukkan nama barang yang dicari: ")

ditemukan = False

for item in barang:
    if item.lower() == cari.lower():
        ditemukan = True
        break

if ditemukan:
    print("Barang ditemukan")
else:
    print("Barang tidak ditemukan")
```

##Studi Kasus 3: Pengurutan Nilai
-> Deskripsi: Dosen ingin mengurutkan nilai mahasiswa dari kecil ke besar

```python
nilai = [85, 70, 95, 60, 80]

print("Data Sebelum Diurutkan:")
print(nilai)

nilai.sort()

print("Data Setelah Diurutkan:")
print(nilai)
```

##Studi Kasus 4 - Update Data Mahasiswa
-> Deskripsi: Sistem akademik ingin memperbarui data mahasiswa

```python
mahasiswa = ["Kit", "Lili", "Petter", "Dorothy", "Brien"]

print("Data Sebelum Update:")
print(mahasiswa)

mahasiswa[2] = "Richard"

print("Data Setelah Update:")
print(mahasiswa)
```

##DISCUSSION

Soal 1: Apa perbedaan array dan list

=> Jawab: Array dan list sama-sama digunakan untuk menyimpan banyak data dalam satu variabel. Namun, pada Python yang paling sering digunakan adalah list karena lebih fleksibel. List dapat menyimpan berbagai tipe data dalam satu tempat dan ukurannya dapat diubah sesuai kebutuhan. Sementara itu, array umumnya memiliki tipe data yang sama untuk setiap elemennya dan lebih terstruktur.

Soal 2: Mengapa list disebut struktur data linear

=> Jawab: List disebut struktur data linear karena data di dalamnya disusun secara berurutan dari elemen pertama hingga elemen terakhir. Setiap elemen memiliki posisi atau indeks tertentu sehingga dapat diakses satu per satu sesuai urutannya. Oleh karena itu, hubungan antar data dalam list bersifat berurutan atau linear.

Soal 3: Apa fungsi transversal

=> Jawab: Traversal berfungsi untuk menelusuri atau membaca seluruh elemen yang ada dalam suatu list. Proses ini biasanya dilakukan menggunakan perulangan seperti for. Traversal sangat penting karena menjadi dasar berbagai operasi pengolahan data, seperti menampilkan data, mencari data, menghitung jumlah data, maupun melakukan analisis terhadap data yang tersimpan.

Soal 4: Mengapa searching penting dalam pengolahan data

=> Jawab: Searching penting karena membantu pengguna menemukan data yang dibutuhkan dengan lebih cepat dan mudah. Dalam pengolahan data, sering kali diperlukan pencarian informasi tertentu dari sekumpulan data yang tersedia. Tanpa proses searching, pengguna harus memeriksa data satu per satu secara manual yang tentu kurang efisien, terutama jika jumlah datanya banyak.

Soal 5: Apa keuntungan sorting data

=> Jawab: Sorting atau pengurutan data membuat data menjadi lebih rapi dan terstruktur. Data yang sudah terurut akan lebih mudah dibaca, dianalisis, dan dicari. Selain itu, beberapa algoritma pencarian dapat bekerja lebih cepat jika data sudah berada dalam kondisi terurut. Oleh karena itu, sorting sering digunakan sebagai langkah awal dalam proses pengolahan data.
