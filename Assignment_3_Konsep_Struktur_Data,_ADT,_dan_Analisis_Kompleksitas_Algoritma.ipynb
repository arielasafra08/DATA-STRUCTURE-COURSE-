##Tugas 2 - Simulasi Stack Sederhana

```python
stack = []

stack.append("Buku A")
stack.append("Buku B")
stack.append("Buku C")

print("Isi Stack:", stack)

data_keluar = stack.pop()

print("Data yang keluar:", data_keluar)
print("Isi Stack sekarang:", stack)
```

> Langkah Praktikum

- Membuat list kosong sebagai stack.
- Menambahkan beberapa data menggunakan append().
- Mengambil data menggunakan pop().
- Menampilkan hasil sebelum dan sesudah pengambilan data.

> Analisis

Program mensimulasikan ADT Stack menggunakan list Python. Data yang terakhir dimasukkan akan menjadi data pertama yang keluar ketika operasi pop dijalankan. Hal ini menunjukkan konsep LIFO (Last In First Out). Operasi push dan pop dapat dilakukan dengan mudah menggunakan fungsi bawaan Python.

> Kesimpulan

Stack bekerja berdasarkan prinsip LIFO, yaitu data yang terakhir masuk akan menjadi data pertama yang keluar. Konsep ini banyak digunakan pada sistem navigasi, undo, dan manajemen memori.

##Tugas 3 - Tranversal Data menggunakan Perulangan

```python
data = [15, 25, 35, 45, 55]

for item in data:
    print(item)
```

> Langkah Praktikum

- Menggunakan perulangan for.
- Menampilkan setiap elemen list.

Traversal adalah proses mengunjungi seluruh elemen dalam suatu struktur data. Pada program ini setiap elemen pada list dibaca satu per satu menggunakan perulangan. Jumlah proses yang dilakukan sama dengan jumlah data yang ada dalam list.

> Kesimpulan

Traversal merupakan operasi dasar yang sering digunakan dalam pengolahan data karena memungkinkan seluruh data diproses secara sistematis.

##Tugas 4 - Menghitung Jumlah Iterasi dalam Loop

```python
jumlah_iterasi = 0

for i in range(10):
    print("Iterasi ke-", i + 1)
    jumlah_iterasi += 1

print("Total Iterasi:", jumlah_iterasi)
```

> Langkah Praktikum

- Membuat perulangan sebanyak 10 kali.
- Menambahkan variabel penghitung iterasi.
- Menampilkan jumlah iterasi setelah perulangan selesai.
- Mengamati hasil yang diperoleh.

> Analisis

Program menghitung jumlah pengulangan yang terjadi selama proses looping. Setiap kali loop berjalan, nilai variabel penghitung bertambah satu. Dari hasil percobaan terlihat bahwa jumlah iterasi sama dengan jumlah perulangan yang dilakukan. Program ini memiliki kompleksitas O(n) karena jumlah proses bertambah sebanding dengan jumlah data.

> Kesimpulan

Jumlah iterasi dapat digunakan untuk mengetahui banyaknya proses yang dilakukan oleh suatu algoritma dan membantu memahami konsep kompleksitas waktu.

##Tugas 5 - Nested Loop dan Analisis Kompleksitas

```python
for i in range(5):
    for j in range(5):
        print(i, j)
```

> Langkah Praktikum:

- Membuat perulangan bersarang (nested loop).
- Menjalankan program.
- Mengamati jumlah output yang dihasilkan.
- Menganalisis kompleksitas algoritma.

> Analisis

Program menggunakan dua perulangan yang saling bersarang. Loop pertama berjalan sebanyak 5 kali dan untuk setiap iterasi loop pertama, loop kedua juga berjalan sebanyak 5 kali. Total proses yang terjadi adalah 25 kali. Oleh karena itu kompleksitas algoritmanya termasuk O(n²).

> Kesimpulan

Nested loop menghasilkan jumlah proses yang lebih besar dibandingkan single loop sehingga membutuhkan waktu eksekusi yang lebih lama ketika data semakin banyak.

##Tugas 6 - Simulasi Pencarian Data Sederhana

```python
data = [10, 20, 30, 40, 50]
target = 40

for item in data:
    if item == target:
        print("Data ditemukan")
        break
```

> Langkah Praktikum

- Membuat list data.
- Menentukan data yang akan dicari.
- Menggunakan perulangan untuk memeriksa setiap elemen.
- Menampilkan hasil pencarian.

> Analisis

Program menggunakan metode Linear Search untuk mencari data tertentu dalam list. Proses pencarian dilakukan dengan memeriksa setiap elemen secara berurutan hingga data ditemukan. Jika data ditemukan maka perulangan dihentikan menggunakan perintah break.

> Kesimpulan

Linear Search merupakan algoritma pencarian yang sederhana dan mudah dipahami, tetapi kurang efisien jika jumlah data sangat besar.

##Tugas 7 - Perbandingan Dua Algoritma Pencarian

```python
data = [10, 20, 30, 40, 50]

# Linear Search
target = 40
for item in data:
    if item == target:
        print("Linear Search: Data ditemukan")
        break

# Pencarian menggunakan operator in
if target in data:
    print("Operator in: Data ditemukan")
```

> Langkah Praktikum

- Menyiapkan data dalam list.
- Melakukan pencarian menggunakan Linear Search.
- Melakukan pencarian menggunakan operator in.
- Membandingkan hasil keduanya.

> Analisis

Kedua metode menghasilkan output yang sama yaitu menemukan data yang dicari. Namun operator in lebih ringkas dalam penulisan kode karena proses pencarian dilakukan secara otomatis oleh Python. Walaupun demikian, konsep dasar pencarian tetap perlu dipahami agar dapat menganalisis efisiensi algoritma.

> Kesimpulan

Terdapat berbagai cara untuk melakukan pencarian data. Pemilihan metode yang tepat dapat mempermudah penulisan program dan meningkatkan efisiensi proses pencarian.

##Tugas 8 - Simulasi Sistem Antrian Mahasiswa

```python
antrian = []

antrian.append("Mahasiswa 1")
antrian.append("Mahasiswa 2")
antrian.append("Mahasiswa 3")

print("Antrian Saat Ini:")
print(antrian)

layani = antrian.pop(0)

print("Mahasiswa yang dilayani:", layani)
print("Sisa Antrian:", antrian)
```

> Langkah Praktikum

- Membuat list sebagai antrian.
- Menambahkan beberapa mahasiswa ke dalam antrian.
- Mengambil mahasiswa yang berada di urutan pertama.
- Menampilkan kondisi antrian setelah pelayanan.

> Analisis

Program mensimulasikan konsep Queue atau antrian. Data yang pertama masuk akan menjadi data pertama yang keluar. Konsep ini disebut FIFO (First In First Out). Sistem antrian banyak digunakan pada pelayanan pelanggan, rumah sakit, bank, dan sistem tiket.

> Kesimpulan

Queue merupakan struktur data yang bekerja berdasarkan prinsip FIFO sehingga cocok digunakan untuk mengelola proses pelayanan yang memerlukan urutan kedatangan.

##Tugas 9 - Analisis Kompleksitas Algoritma

```python
# O(1)
data = [10, 20, 30]
print(data[0])

# O(n)
for item in data:
    print(item)

# O(n²)
for i in range(3):
    for j in range(3):
        print(i, j)
```

> Langkah Praktikum

- Membuat contoh algoritma dengan kompleksitas berbeda.
- Menjalankan program.
- Mengamati jumlah proses yang terjadi.
- Membandingkan tingkat efisiensinya.

> Analisis

Bagian pertama memiliki kompleksitas O(1) karena hanya mengakses satu elemen. Bagian kedua memiliki kompleksitas O(n) karena jumlah proses mengikuti jumlah data. Bagian ketiga memiliki kompleksitas O(n²) karena terdapat nested loop yang menyebabkan jumlah proses meningkat lebih cepat. Semakin besar nilai kompleksitas, semakin besar pula waktu yang dibutuhkan program untuk menyelesaikan proses.

> Kesimpulan

Kompleksitas algoritma digunakan untuk mengukur efisiensi program. Algoritma dengan kompleksitas lebih kecil umumnya memiliki performa yang lebih baik dibandingkan algoritma dengan kompleksitas yang lebih besar, terutama ketika jumlah data terus bertambah.
