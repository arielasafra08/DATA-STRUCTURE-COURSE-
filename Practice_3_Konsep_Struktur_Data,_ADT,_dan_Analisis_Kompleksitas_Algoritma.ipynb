##Dasar Teori
> Struktur Data

Cara untuk menyimpan, mengorganisasikan, dan mengelola data di dalam komputer sehingga data dapat digunakan secara efisien

> Jenis Struktur Data

Struktur data linear (array, list, stact, queue), struktur data non linear (tree, graph)

> Abstract Data Type (ADT)

Model konseptual yang mendefinisikan jenis data, operasi yang dapat dilakukan, dan perilaku struktur data

##Studi Kasus 2: Simulasi Stack Buku
-> Deskripsi: Tumpukan buku di perpustakaan mengikuti konsep LIFO

```python
stack_buku = []

stack_buku.append("The Psychology of Money")
stack_buku.append("The Courage to be Disliked")
stack_buku.append("The 5 am Club")

print("Tumpukan Buku Saat Ini:")
print(stack_buku)

buku_diambil = stack_buku.pop()

print("Buku yang Diambil:", buku_diambil)
print("Sisa Tumpukan Buku:")
print(stack_buku)
```

##Studi Kasus 3: Analisis Perulangan
-> Deskripsi: Dosen ingin menunjukkan perbedaan jumlah proses pada loop biasa dan nested loop

```python
# Single loop

data = [1, 2, 3, 4, 5]

for i in data:
    print(i)
```

```python
# Nested loop

for i in range(5):
    for j in range(5):
        print(i, j)
```

##Studi Kasus 4: Pencarian Data
-> Deskripsi: Sebuah koperasi ingin mencari ID anggota tertentu

```python
id_anggota = [101, 102, 103, 104, 105]
target = int(input("Masukkan ID Anggota yang dicari: "))

ditemukan = False

for data in id_anggota:
    if data == target:
        ditemukan = True
        break

if ditemukan:
    print("ID Anggota ditemukan")
else:
    print("ID Anggota tidak ditemukan")
```

##DISCUSSION

Soal 1: Mengapa struktur data penting dalam pemrograman

=> Jawab: Struktur data penting karena digunakan untuk mengatur dan menyimpan data agar lebih terorganisir. Dengan struktur data yang tepat, proses pengolahan data menjadi lebih cepat, efisien, dan mudah dikelola. Selain itu, struktur data juga membantu programmer membuat program yang lebih rapi dan mudah dikembangkan.

Soal 2: Apa perbedaan struktur data linear dan non linear

=> Jawab: Struktur data linear adalah struktur data yang elemen-elemennya tersusun secara berurutan sehingga setiap data memiliki hubungan langsung dengan data sebelum dan sesudahnya. Contohnya adalah array, list, stack, dan queue.

Sedangkan struktur data non-linear adalah struktur data yang penyusunan datanya tidak berurutan, melainkan berbentuk hierarki atau jaringan. Contohnya adalah tree dan graph. Struktur data non-linear biasanya digunakan untuk menangani hubungan data yang lebih kompleks.

Soal 3: Mengapa algoritma harus efisien

=> Jawab: Algoritma harus efisien agar program dapat berjalan lebih cepat dan menggunakan sumber daya komputer secara optimal. Jika algoritma tidak efisien, waktu eksekusi akan semakin lama ketika jumlah data bertambah besar. Oleh karena itu, efisiensi algoritma sangat penting untuk meningkatkan kinerja program dan memberikan hasil yang lebih baik.

Soal 4: Apa fungsi Big-O Notation

=> Jawab: Big-O Notation digunakan untuk mengukur dan menggambarkan tingkat kompleksitas suatu algoritma. Dengan Big-O, kita dapat mengetahui seberapa besar pengaruh jumlah data terhadap waktu eksekusi program. Selain itu, Big-O membantu membandingkan efisiensi beberapa algoritma sehingga dapat dipilih algoritma yang paling optimal untuk menyelesaikan suatu masalah.

Soal 5: Mengapa nested loop lebih lambat dibanding single loop

=> Jawab: Nested loop lebih lambat karena terdapat perulangan di dalam perulangan. Akibatnya jumlah proses yang dilakukan menjadi jauh lebih banyak dibandingkan single loop. Jika single loop memiliki kompleksitas O(n), maka nested loop umumnya memiliki kompleksitas O(n²). Semakin besar jumlah data, semakin besar pula perbedaan waktu eksekusi antara keduanya, sehingga nested loop membutuhkan waktu pemrosesan yang lebih lama.
