##Tugas 7 - Pengelolaan Nilai Mahasiswa

```python
nilai = []

for i in range(5):
    data = int(input(f"Masukkan nilai ke-{i+1}: "))
    nilai.append(data)

rata_rata = sum(nilai) / len(nilai)
tertinggi = max(nilai)

print("Daftar Nilai :", nilai)
print("Rata-rata    :", rata_rata)
print("Nilai Tertinggi:", tertinggi)
```

Program menerima lima input nilai dari pengguna dan menyimpannya ke dalam list. Fungsi sum() digunakan untuk menghitung total nilai, kemudian dibagi jumlah data untuk memperoleh rata-rata. Nilai tertinggi diperoleh menggunakan fungsi max().

> Kesimpulan

List sangat membantu dalam pengolahan data numerik karena memungkinkan perhitungan statistik sederhana seperti rata-rata dan nilai tertinggi.

##Tugas 9 - Sorting Manual tanpa menggunakan Sort()

```python
data = [45, 12, 78, 23, 56]

for i in range(len(data)):
    for j in range(i + 1, len(data)):
        if data[i] > data[j]:
            data[i], data[j] = data[j], data[i]

print("Hasil Sorting:", data)
```

> Langkah Praktikum

Membuat list angka acak.
Membandingkan setiap elemen.
Menukar posisi data jika diperlukan.
Menampilkan hasil pengurutan.

> Analisis

Program menggunakan metode sorting sederhana dengan membandingkan setiap elemen terhadap elemen lainnya. Jika ditemukan nilai yang lebih kecil, maka posisi kedua data ditukar. Proses ini berlangsung hingga seluruh data tersusun dari kecil ke besar.

> Kesimpulan

Sorting manual membantu memahami konsep dasar algoritma pengurutan sebelum menggunakan fungsi bawaan Python seperti sort() atau sorted().
